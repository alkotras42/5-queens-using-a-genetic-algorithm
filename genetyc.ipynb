```python
import random as rd
import time
from math import *
```

```python
def create_population(pop_size):
    population = []
    for i in range(pop_size):
        cond = True
        arr = []
        while cond:
            for j in range(8):
                a = rd.randint(0,8)
                arr.append(a)
            if arr.count(0) != 3:
                arr = []
            else:
                cond = False
        population.append(arr)
    return population
```

```python
def create_next_popolation(population, mute_rate):
    Mating_pool = []
    for i in range(len(population)):
        n = calc_fitnes(population[i]) * 5
        for j in range(n):
            Mating_pool.append(population[i])
    for i in range(len(population)): 
        p1 = rd.choice(Mating_pool)
        p2 = rd.choice(Mating_pool)
        child = crossover(p1, p2)
        child = make_zeros(child)
        child = mutation(child, mute_rate)
        population[i] = child
    return population
```

```python
def calc_max_fitnes(population):
    max_fitnes = 0
    for i in range(len(population)):
        if calc_fitnes(population[i]) > max_fitnes:
            max_fitnes = calc_fitnes(population[i])
            num = i
    return max_fitnes, num
```

```python
def make_zeros(pop):
    cond = True
    while cond:
        if pop.count(0) != 3:
            if pop.count(0) < 3:
                for j, val in enumerate(pop):
                    if val > 0:
                        pop[j] = 0
                        break
            if pop.count(0) > 3:
                for j, val in enumerate(pop):
                    if val == 0:
                        pop[j] = rd.randint(1,8)
                        break
        else:
            cond = False
    return pop
```

```python
def calc_fitnes(pop):
    fitnes = 0
    for i in range(len(pop)):
        for j in range(len(pop)):
            for k, val in enumerate(pop):
                x1 = i + 1
                y1 = j + 1
                x2 = val 
                y2 = k + 1
                if (val != 0) and ((abs(x1 - x2) == abs(y1 - y2)) or (x1 == x2) or (y1 == y2)):
                    fitnes +=1
                    break
    return fitnes
```

```python
def mutation(pop, mute_rate):
    for i in range(len(pop)):
        r = rd.random()
        if (r < mute_rate) and (pop[i] != 0):
            pop[i] = rd.randint(1,8)
    return pop
```

```python
def crossover(p1, p2):
    r = rd.randint(0,8)
    child = p1[:r] + p2[r:]
    return child
```

```python
def is_finish(pop):
    if calc_fitnes(pop) == 28:
        return True
    else:
        return False
```

```python
def print_pop(pop):
    start_str = '--------'
    res = []
    for i in range(8):
        res.append(''.join(start_str))
    for i in range(len(res)+1):
        for j, val in enumerate(pop,1):
            if (i == val) and (val > 0):
                temp_str = list(res[i-1])
                temp_str[j-1] = '*'
                res[i-1] = temp_str
                res[i-1] = ''.join(res[i-1])
    res = '\n'.join(res)
    for i in res:
        return (res)
```

```python
def main():
    pop_size = 10
    mute_rate = 0.05
    new_population = create_population(pop_size)
    a = rd.choice(new_population)
    b = rd.choice(new_population)
    loop = True
    res = 0
    it = 0
    while (loop):
        new_population = create_next_popolation(new_population, mute_rate)
        res = calc_max_fitnes(new_population)[0]
        it += 1
        print('{0}: {1}'.format(it, res))
        if res == 64:
            a = new_population[calc_max_fitnes(new_population)[1]]
            print(a)
            print(print_pop(a))
            loop = False
```
